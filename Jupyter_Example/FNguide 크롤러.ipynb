```python
import requests
```

```python
headers = {
    "key": "B8DE27145B59CF9845FE",
    "format": "json",
    "success": "true",
    "errcd": "",
    "errmsg": "",
    "code": "A005930,A055550,A009200,A024110,A253590",
    "item": "M111000,M113000,M122700,M221100,M312000,M314000,M382100",
    "consolgb": "M",
    "annualgb": "QQ",
    "accdategb": "C",
    "fraccyear": "201203",
    "toaccyear": "202209"
}
```

```python
url = "https://www.fnspace.com/Api/FinanceApi"
res = requests.get(url,headers)
res
```

```python
resdict = res.json()
```

```python
resdict['dataset'][0]
```

```python
resarr = []
for ele in resdict['dataset']:
    for d in ele['DATA']:
        d['CODE'] = ele['CODE'][1:]
        resarr.append(d)
```

```python
import pandas as pd
df = pd.DataFrame(resarr)
```

```python
df
```

```python
df = df.set_index([pd.to_datetime(df['DATE'].apply(lambda x : x + '01')),df['CODE']])
```

```python
df.drop(['DATE','YYMM','FS_YEAR','FS_MONTH','FS_QTR','MAIN_TERM'],axis=1,inplace=True)
```
